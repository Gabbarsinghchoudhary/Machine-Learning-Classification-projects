```python
# Dataset Preprocessing
```

```python
# First, I'll do the preprocessing of data
#     1. Convert categorical data to numeric data
#     2. Replace values in class column:
#         * 1 for pisonous
#         * 0 for Edible
```

```python
not_change = ['class']
change_cols = list(set(mushroom.columns) - set(not_change))
df = pd.get_dummies(mushroom, columns=change_cols)
```

```python
df.tail()
```

```python
# Replacing values in class column
df['class'] = df['class'].replace('p', 1)
df['class'] = df['class'].replace('e', 0)
```

```python

mushroom['stalk-root'].unique()
len(mushroom.columns)
for i in range(0, len(mushroom.columns)):
    print("{}  =  {} \n".format(mushroom.columns[i],mushroom[ mushroom.columns[i]].unique()))
```

```python
df.head()
```

```python
# Shuffle the data
df = df.sample(frac=1)
df.head()
```

```python
df.corr()
```

```python

sns.heatmap(df)
```

```python
# Dataset Training/Classifying Algorithms
```

```python
# Predicting using KNeighoborsClassifier
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

# Set up random seed
np.random.seed(42)
```

```python
# Create the data
X = df.drop('class',axis=1)
y = df["class"]
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

```python
# Instantiate the model
clf = KNeighborsClassifier()

# Fit
clf.fit(X_train, y_train)

# Score
clf.score(X_test, y_test)
```

```python
print(np.array(y_test[:5]))
```

```python
# Make predictions
y_pred = clf.predict(X_test)


chack_pred_df = pd.DataFrame({"Actual_values": np.array(y_test),
                      "Predicted_values": np.array(y_pred)})
chack_pred_df.head()
```

```python
# pisonous
# Edible
chack_pred_df['Actual_values'] = chack_pred_df['Actual_values'].replace(0,'Edible')
chack_pred_df['Actual_values'] = chack_pred_df['Actual_values'].replace(1,'Pisonous')
chack_pred_df['Predicted_values'] = chack_pred_df['Actual_values'].replace(0,'Edible')
chack_pred_df['Predicted_values'] = chack_pred_df['Actual_values'].replace(1,'Pisonous')
```

```python
chack_pred_df.head()
```

```python
# Confusion matrix and classification report
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(chack_pred_df['Actual_values'], chack_pred_df['Predicted_values'])
```

```python
def plot_conf_mat(test_value, predict_value):
    conf_mat = confusion_matrix(test_value, predict_value)
    plt.figure(figsize=(4,4))
    sns.heatmap(conf_mat, annot=True, cbar=False)
    
    plt.xlabel("Predicted Label")
    plt.ylabel("Actual Label")    
    plt.show()
```

```python
confusion_matrix(y_test, y_pred)
```

```python
plot_conf_mat(chack_pred_df['Actual_values'], chack_pred_df['Predicted_values'])
```

```python
print(classification_report(y_test, y_pred))
```

```python
# Predicting using RandomForestClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier

# Set up random seed
np.random.seed(42)

# Create data

X = df.drop('class', axis=1)
y = df['class']

# Split the data
X_train,X_test, y_train,y_test = train_test_split(X, y, test_size=.2)


# INstantiate the model
clf = RandomForestClassifier(max_depth=8, random_state=0)

# Fit the model
clf.fit(X_train, y_train)

# Score
clf.score(X_test, y_test)
```

```python
# Make predictions
y_pred = clf.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
# Confusion matrix
plot_conf_mat(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
